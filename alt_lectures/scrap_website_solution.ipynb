### Exercise

Look at this site: https://www.cs.uu.nl/education that lists all courses offered by our Dept. of Computing Science. Scrap this website to obtain a list of all course codes.

Check the documentation of Beautiful Soup here: https://www.crummy.com/software/BeautifulSoup/bs4/doc/

Some functions/attributes you might find useful are: `find_all()` and `contents`.

```python
import bs4
import requests

# GET document from the website:
response = requests.get("https://www.cs.uu.nl/education/")

# get HMTL content from response object
html_doc = response.text

# turn into BS4 object
soup = bs4.BeautifulSoup(html_doc, 'html.parser')

print(soup.title)

# get the course-table; it is the 3rd table in the webpage:
courseTable = soup.find_all("table")[2]
# this gives all the rows in the table:
rows = courseTable.find_all("tr")
# this print all the course codes:
for row in rows:
    print(row.contents[2].string)
```
